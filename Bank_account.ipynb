```python
import datetime
```

```python
class BankAccount:

  def __init__(self,name):
    self.name = name
    self.accountBalance = 0

  def deposit(self,amount):
    if amount <= 0:
            print("An amount less or equal to zero cannot be deposited ")
            return

    current_time = datetime.datetime.now()

    self.accountBalance += amount
    print(f"Your account balance is {self.accountBalance}. Transaction: Deposit. Date and time: {current_time.strftime('%Y-%m-%d %H:%M:%S')}")


  def withdraw(self, amount):
    if amount <= 0:
            print("An amount less or equal to zero cannot be withdraw ")
            return
    current_time = datetime.datetime.now()
    if self.accountBalance >= amount:
      self.accountBalance -= amount
      print(f"Your account balance is {self.accountBalance}. Transaction: Withdrawal. Date and time: {current_time.strftime('%Y-%m-%d %H:%M:%S')}")
    else:
      print(f"Insufficient funds. Your account balance is {self.accountBalance}")



  def show_balance(self):
    print(f"Your account balance is {self.accountBalance}")



class SavingAccount(BankAccount):
    def __init__(self, name):
        super().__init__(name)
    def deposit(self, amount):  # Polymorphism 
        current_time = datetime.datetime.now()

        bonus = amount * 0.05
        self.accountBalance += amount + bonus
        print(
            f"Deposited {amount} + bonus {bonus}. "
            f"Date and time: {current_time.strftime('%Y-%m-%d %H:%M:%S')}"
        )




```

```python
user1= BankAccount("Mooj")
user1.deposit(2000)
user1.withdraw(150)
user1.show_balance()

user2= SavingAccount("Mooj")
user2.deposit(2000)
user2.show_balance()
```
